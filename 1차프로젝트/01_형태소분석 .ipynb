```python
from konlpy.tag import Okt
from collections import Counter

def analyze_text(text):
    okt = Okt()
    morphs = okt.pos(text, stem=True, join=True)
    return [morph.split('/')[0] for morph in morphs if morph.split('/')[1] not in ['Alpha', 'Number', 'Punctuation', 'Foreign', 'KoreanParticle', 'Josa']]

# 미리보기용 데이터프레임
# df_good = df['장점'].iloc[0:4]
# df_bad = df['단점'].iloc[0:4]
df['장점']=df['장점'].apply(lambda x: analyze_text(str(x)))
df['단점']=df['단점'].apply(lambda x: analyze_text(str(x)))
```

```python
from konlpy.tag import Okt
import re

okt = Okt()

# 정규표현식을 사용하여 문자열 변환하는 함수 정의
def apply_regular_expression(text):
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣\W]')
    result = hangul.sub('', text)
    return result

def remove(text):
    # 쉼표(,) 제거
    cleaned_text = text.replace(',', '')
    # 개행문자(\n) 제거
    cleaned_text = cleaned_text.replace('\n', '')
    return cleaned_text

# 데이터프레임의 모든 열에 함수 적용하여 변환
df['장점'] = df['장점'].apply(lambda x: apply_regular_expression(x) if isinstance(x, str) else x)
df['단점'] = df['단점'].apply(lambda x: apply_regular_expression(x) if isinstance(x, str) else x)

df['장점'] = df['장점'].apply(lambda x: remove(x) if isinstance(x, str) else x)
df['단점'] = df['단점'].apply(lambda x: remove(x) if isinstance(x, str) else x)

# 결과 확인
print(df)
```

```python
def analyze_text(df, stopwords):
    okt = Okt()
    for column in df.columns:
        df[column] = df[column].apply(lambda row: ' '.join([morph for morph, pos in okt.pos(str(row), stem=True, norm=True) if pos not in ['Alpha', 'Number', 'Punctuation', 'Foreign', 'KoreanParticle', 'Josa'] and morph not in stopwords]))

# 불용어(stopwords) 설정
# 불용어 정의
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다', '되다','것']
# stopwords = pd.read_csv("https://raw.githubusercontent.com/yoonkt200/FastCampusDataset/master/korean_stopwords.txt").values.tolist()

# analyze_text 함수 호출
analyze_text(df, stopwords)

# 결과 확인
print(df)
```
